```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

folder_path = r"/kaggle/input/datasets/behrad3d/nasa-cmaps/CMaps"

datasets = ['FD001', 'FD002', 'FD003', 'FD004']

columns = (
    ['engineId', 'cycle',
     'setting_1', 'setting_2', 'setting_3']
    + [f'sensor_{i}' for i in range(1, 22)]
)

print("Columns:")
print(columns)

all_data = {}

for ds in datasets:

    print("\n" + "=" * 60)
    print(f"Processing {ds}...")
    print("=" * 60)

    file_path = os.path.join(folder_path, f"train_{ds}.txt")

    df = pd.read_csv(file_path, sep=r"\s+", header=None)

    # Remove completely empty columns (if any)
    df = df.dropna(axis=1, how='all')

    # Assign column names
    df.columns = columns

    # Store dataset
    all_data[ds] = df

    print("\nShape:", df.shape)
    print("Number of Duplicates:", df.duplicated().sum())
    print("Number of Engines:", df['engineId'].nunique())
    print("Maximum Cycle:", df['cycle'].max())
    print("Missing Values:", df.isnull().sum().sum())

    print("\nStatistical Summary:")
    print(df.describe().T)

    print("\nTail of the Dataset:")
    print(df.tail())

    print("\nInfo of the Dataset:")
    df.info()

    max_cycle = df.groupby('engineId')['cycle'].transform('max')
    df['RUL'] = max_cycle - df['cycle']

    print("\nRUL column added successfully.")
    print(df[['engineId', 'cycle', 'RUL']].head())

    print("\nCorrelation Matrix:")
    print(df.corr(numeric_only=True))

    engine_life = df.groupby('engineId')['cycle'].max()

    plt.figure(figsize=(8, 5))
    plt.hist(engine_life, bins=20, edgecolor='black')
    plt.title(f"{ds} - Engine Life Distribution")
    plt.xlabel("Maximum Cycles")
    plt.ylabel("Number of Engines")
    plt.show()

    max_time_cycles = (
        df.groupby('engineId')['cycle']
        .max()
        .reset_index()
    )

    sns.displot(
        max_time_cycles['cycle'],
        kde=True,
        bins=20,
        height=6,
        aspect=2
    )
    plt.xlabel("Maximum Time Cycles")
    plt.show()

    plt.figure(figsize=(8, 4))
    sns.violinplot(x=max_time_cycles['cycle'].values)
    plt.title(f"{ds} - Violin Plot of Maximum Cycles")
    plt.show()

    plt.figure(figsize=(8, 4))
    sns.boxplot(x=max_time_cycles['cycle'])
    plt.title(f"{ds} - Box Plot of Maximum Cycles")
    plt.show()

    plt.figure(figsize=(8, 4))
    sns.kdeplot(x=max_time_cycles['cycle'], fill=True)
    plt.title(f"{ds} - KDE Plot of Maximum Cycles")
    plt.show()

    plt.figure(figsize=(20, 20))
    sns.heatmap(
        df.corr(numeric_only=True),
        annot=True,
        cmap='Spectral'
    )
    plt.title(f"{ds} - Correlation Heatmap")
    plt.show()
```

```python

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

datasets = ['FD001', 'FD002', 'FD003', 'FD004']

data_path = "/kaggle/input/datasets/behrad3d/nasa-cmaps/CMaps"

columns = (
    ['unit_number', 'time_in_cycles'] +
    [f'op_setting_{i}' for i in range(1, 4)] +
    [f'sensor_{i}' for i in range(1, 22)]
)

test_data = {}
rul_data = {}

for ds in datasets:
    test_file = os.path.join(data_path, f"test_{ds}.txt")
    rul_file = os.path.join(data_path, f"RUL_{ds}.txt")

    test_df = pd.read_csv(
        test_file,
        sep=r"\s+",
        header=None
    )

    test_df = test_df.iloc[:, :26]
    test_df.columns = columns
    test_data[ds] = test_df

    rul_df = pd.read_csv(
        rul_file,
        sep=r"\s+",
        header=None,
        names=['RUL']
    )

    rul_data[ds] = rul_df

output_dir = "test_eda_outputs"
os.makedirs(output_dir, exist_ok=True)

summary = []

for ds in datasets:
    print("\n" + "=" * 60)
    print(f"PROCESSING TEST DATASET: {ds}")
    print("=" * 60)

    test_df = test_data[ds].copy()
    rul_df = rul_data[ds].copy()

    print(f"Shape: {test_df.shape}")
    print(f"Number of Engines: {test_df['unit_number'].nunique()}")
    print(f"Maximum Cycle: {test_df['time_in_cycles'].max()}")
    print(f"Missing Values: {test_df.isnull().sum().sum()}")

    final_cycles = (
        test_df.groupby('unit_number')['time_in_cycles']
        .max()
        .reset_index(name='last_observed_cycle')
    )

    rul_df = rul_df.reset_index(drop=True)
    rul_df['unit_number'] = rul_df.index + 1

    final_cycles = pd.merge(
        final_cycles,
        rul_df[['unit_number', 'RUL']],
        on='unit_number',
        how='left'
    )

    final_cycles.rename(columns={'RUL': 'True_RUL'}, inplace=True)

    final_cycles['Estimated_Failure_cycle'] = (
        final_cycles['last_observed_cycle']
        + final_cycles['True_RUL']
    )

    final_cycles.to_csv(
        os.path.join(output_dir, f"{ds}_final_cycles.csv"),
        index=False
    )

    summary.append({
        'Dataset': ds,
        'Rows': test_df.shape[0],
        'Engines': test_df['unit_number'].nunique(),
        'RUL Records': len(rul_df),
        'Mean True RUL': final_cycles['True_RUL'].mean(),
        'Min True RUL': final_cycles['True_RUL'].min(),
        'Max True RUL': final_cycles['True_RUL'].max()
    })

    plt.figure(figsize=(8, 4))
    sns.histplot(final_cycles['True_RUL'], bins=30, kde=True)
    plt.title(f"{ds} - True RUL Distribution")
    plt.xlabel("RUL")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
    plt.close()

    plt.figure(figsize=(8, 4))
    plt.hist(
        final_cycles['last_observed_cycle'],
        bins=20,
        edgecolor='black'
    )
    plt.title(f"{ds} - Engine Life Distribution (Test Data)")
    plt.xlabel("Maximum Cycles Observed")
    plt.ylabel("Number of Engines")
    plt.tight_layout()
    plt.show()
    plt.close()

    fig, axes = plt.subplots(1, 3, figsize=(18, 4))

    sns.histplot(
        final_cycles['last_observed_cycle'],
        kde=True,
        bins=20,
        ax=axes[0]
    )
    axes[0].set_title(f"{ds} - Max Time Cycles Dist")

    sns.violinplot(
        x=final_cycles['last_observed_cycle'].values,
        ax=axes[1]
    )
    axes[1].set_title(f"{ds} - Violin Plot of Max Cycles")

    sns.boxplot(
        x=final_cycles['last_observed_cycle'],
        ax=axes[2]
    )
    axes[2].set_title(f"{ds} - Box Plot of Max Cycles")

    plt.tight_layout()
    plt.show()
    plt.close()

    plt.figure(figsize=(12, 10))
    sns.heatmap(
        test_df.corr(numeric_only=True),
        annot=False,
        cmap='Spectral'
    )
    plt.title(f"{ds} - Correlation Heatmap (Test Data)")
    plt.tight_layout()
    plt.show()
    plt.close()

print("\n" + "=" * 80)
print("ALL TEST DATASETS SUMMARY")
print("=" * 80)

summary_df = pd.DataFrame(summary)
print(summary_df.to_string(index=False))

summary_df.to_csv(
    os.path.join(output_dir, "all_test_datasets_summary.csv"),
    index=False
)




```
